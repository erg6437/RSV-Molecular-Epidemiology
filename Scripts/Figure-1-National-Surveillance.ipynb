```python
Hosptotal = pd.read_csv("/Users/estefanyguzman/Documents/Northwestern/PhD/Thesis-Work/Clinical-Analysis/RSVNET-Hospitalization.csv")
```

```python
#Converting times to datetime
Hosptotal['Week ending date'] = pd.to_datetime(Hosptotal['Week ending date'])
Hosptotal = Hosptotal[Hosptotal['Week ending date'] > '2018-10-06']
Hosptotal = Hosptotal[(Hosptotal['State'] == 	'Entire Network (RSV-NET)') &(Hosptotal['Sex'] =='Overall') & (Hosptotal['Race'] == 'Overall') & (Hosptotal['MMWR Week'] != 'Overall')]
```

```python
#Stratifying data to designated time ranges set by RSV-Net
#Specify why in the literature why these have been sectioned off this way
months0to4 = Hosptotal.loc[(Hosptotal['Age Category'] == '0-4 years')]
yrs5to17 = Hosptotal.loc[(Hosptotal['Age Category'] == '5-17 years')]
yrs18to49 = Hosptotal.loc[(Hosptotal['Age Category'] == '18-49 years')]
yrs50to64 = Hosptotal.loc[(Hosptotal['Age Category'] == '50-64 years')]
yrs65up = Hosptotal.loc[(Hosptotal['Age Category'] == '65+ years')]
```

```python
import matplotlib.dates as mdates
names = ['0 - 4 years', '5 - 17 years', '18 - 49 years', '50 - 64 years', '65 years and up']
plt.figure(figsize=(10,6))
y= yrs65up['Rate']
plt.plot(months0to4['Week ending date'], months0to4['Rate'], color="black",   markersize = 5, linewidth=0.75)
plt.plot(yrs5to17['Week ending date'], yrs5to17['Rate'], color="maroon",  markersize = 5, linewidth=0.75)
plt.plot(yrs18to49['Week ending date'], yrs18to49['Rate'], color="red",  markersize = 5, linewidth=0.75)
plt.plot(yrs50to64['Week ending date'], yrs50to64['Rate'], color="pink",  markersize = 5, linewidth=0.75)
plt.plot(yrs65up['Week ending date'], yrs65up['Rate'], color="orange",  markersize = 5, linewidth=0.75)


plt.legend(names, loc='right', bbox_to_anchor=(1.3, 0.88),
          ncol=1, fancybox=True, shadow=False, fontsize=12)

plt.yticks(np.arange(min(y), max(y) + 60, 5))
#plt.grid(axis = 'y', ls='dotted') 
plt.ylabel('Hospitalization Rate per 100,000', fontsize=18)
plt.title('RSV-NET Hospitalization Rate in the United States',fontsize=18)
plt.xlabel('Year', fontsize=18)

plt.rcParams['figure.dpi']=1200

# Set the x-axis locator to display years
plt.gca().xaxis.set_major_locator(mdates.YearLocator())

#plt.savefig("RSV-NET_Hosp_rate.eps", format="eps", bbox_inches="tight")
```
